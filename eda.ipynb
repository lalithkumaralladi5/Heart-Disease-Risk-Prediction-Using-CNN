# 🫀 Heart Disease Risk Prediction — Exploratory Data Analysis

This notebook walks through the dataset step-by-step before training.  
Run each cell in order.

**Dataset:** UCI Cleveland Heart Disease  
**Target:** 0 = No Disease (Low Risk), 1 = Disease (High Risk)

```python
# ── Imports ──────────────────────────────────────────────────────────────────
import sys, os
sys.path.insert(0, '..')   # so we can import from utils/

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from utils.preprocessing import load_raw_data, clean_data, FEATURE_COLS

%matplotlib inline
plt.rcParams['figure.dpi'] = 120
print('Imports OK')
```

## 1. Load Data

```python
df_raw = load_raw_data()
print('Shape:', df_raw.shape)
df_raw.head()
```

```python
# Basic info
df_raw.info()
```

```python
# Descriptive statistics
df_raw.describe().T.style.background_gradient(cmap='Blues', subset=['mean', 'std'])
```

## 2. Missing Values

```python
missing = df_raw.isnull().sum()
missing[missing > 0].plot(kind='bar', color='#e53935', edgecolor='black')
plt.title('Missing Value Counts')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
print(missing)
```

## 3. Target Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Count plot
counts = df_raw['target'].value_counts()
axes[0].bar(['Low Risk (0)', 'High Risk (1)'], counts.values,
            color=['#2196F3', '#F44336'], edgecolor='black')
axes[0].set_title('Target Class Distribution')
axes[0].set_ylabel('Count')

# Pie chart
axes[1].pie(counts.values,
            labels=['Low Risk', 'High Risk'],
            autopct='%1.1f%%',
            colors=['#2196F3', '#F44336'],
            startangle=90)
axes[1].set_title('Target Distribution (%)')

plt.tight_layout()
plt.show()
```

## 4. Feature Distributions

```python
from utils.visualization import plot_feature_distribution
plot_feature_distribution(df_raw, FEATURE_COLS, save=False)
```

## 5. Correlation Heatmap

```python
from utils.visualization import plot_correlation_heatmap
plot_correlation_heatmap(df_raw, FEATURE_COLS, save=False)
```

## 6. Clean Data & Verify

```python
df_clean = clean_data(df_raw.copy())
print('Missing after cleaning:', df_clean.isnull().sum().sum())
print('Shape:', df_clean.shape)
```

## 7. Age vs Max Heart Rate (coloured by target)

```python
plt.figure(figsize=(9, 5))
palette = {0: '#2196F3', 1: '#F44336'}
for label, group in df_clean.groupby('target'):
    name = 'Low Risk' if label == 0 else 'High Risk'
    plt.scatter(group['age'], group['thalach'],
                c=palette[label], label=name, alpha=0.7, edgecolors='w', s=60)

plt.xlabel('Age (years)')
plt.ylabel('Max Heart Rate (thalach)')
plt.title('Age vs Max Heart Rate by Disease Status')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
